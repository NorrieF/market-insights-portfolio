```python
import duckdb
from pathlib import Path

Path("data").mkdir(exist_ok=True)
con = duckdb.connect("data/local.duckdb")

con.execute("CREATE TABLE IF NOT EXISTS sanity(x INT);")
con.execute("DELETE FROM sanity;")
con.execute("INSERT INTO sanity VALUES (1), (2), (3);")

con.execute("SELECT COUNT(*) AS n_rows, SUM(x) AS total FROM sanity;").df()
```
